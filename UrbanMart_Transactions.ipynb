# RFM Segmentation

```python
# Determine the last transaction date and add one day to use as a snapshot date
snapshot_date = df['TransactionDate'].max() + dt.timedelta(days=1)

# Group by CustomerID and calculate R, F, and M
rfm = df.groupby('CustomerID').agg({
    'TransactionDate': lambda date: (snapshot_date - date.max()).days,
    'TransactionID': 'count',
    'TransactionValue': 'sum'
})

# Rename the columns to be more descriptive
rfm.rename(columns={'TransactionDate': 'Recency',
                    'TransactionID': 'Frequency',
                    'TransactionValue': 'Monetary'}, inplace=True)

# Display the first few rows of the RFM table
print(rfm.head())
```

```python
# Create RFM scores on a 1-4 scale, where 4 is the best
# For Recency, a lower value is better, so the labels are inverted
r_labels = [4, 3, 2, 1]
f_labels = [1, 2, 3, 4]
m_labels = [1, 2, 3, 4]

# Create the score columns using pd.qcut, which bins the data into quartiles
rfm['R_Score'] = pd.qcut(rfm['Recency'], q=4, labels=r_labels)
rfm['F_Score'] = pd.qcut(rfm['Frequency'], q=4, labels=f_labels)
rfm['M_Score'] = pd.qcut(rfm['Monetary'], q=4, labels=m_labels)

# Combine the scores into a single string for easy segmentation
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)

# Display the first few rows with the new scores
print(rfm.head())
```

```python
# Define a function to segment customers based on their scores
def segment_customer(df):
    if df['RFM_Score'] == '444':
        return 'Best Customers'
    elif df['F_Score'] == 4:
        return 'Loyal Customers'
    elif df['M_Score'] == 4:
        return 'Big Spenders'
    elif df['R_Score'] == 1:
        return 'At Risk'
    elif df['R_Score'] == 4:
        return 'New Customers'
    else:
        return 'Regular Customers'

# Apply the function to the DataFrame to create the 'Segment' column
rfm['Segment'] = rfm.apply(segment_customer, axis=1)

# Display the head of the DataFrame with the new Segment column
print(rfm.head())

# Display the size of each customer segment
print('\nCustomer Segment Sizes:')
print(rfm['Segment'].value_counts())
```

```python
plt.figure(figsize=(10, 6))

# Plot the value counts of the 'Segment' column
segment_counts = rfm['Segment'].value_counts()
ax = sns.barplot(x=segment_counts.values, y=segment_counts.index, palette='viridis')

ax.set_title('Customer Segmentation by RFM Score', fontsize=15, weight='bold')
ax.set_xlabel('Number of Customers', fontsize=12)
ax.set_ylabel('Segment', fontsize=12)

for i, v in enumerate(segment_counts.values):
    ax.text(v + 5, i, str(v), color='black', va='center')

plt.show()
```

```python
sns.set_style("whitegrid")

segment_counts = rfm['Segment'].value_counts()

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=segment_counts.values, y=segment_counts.index, palette='mako')

ax.set_title('Customer Segment Distribution', fontsize=16, weight='bold')
ax.set_xlabel('Number of Customers', fontsize=12)
ax.set_ylabel('Segment', fontsize=12)

for i, v in enumerate(segment_counts.values):
    ax.text(v + 5, i, str(v), color='black', va='center')

plt.show()
```

```python
rfm_corr = rfm[['Recency', 'Frequency', 'Monetary']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(rfm_corr, annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Correlation Heatmap of R, F, and M', fontsize=15, weight='bold')
plt.show()
```

```python
df_merged = df.merge(rfm['Segment'], on='CustomerID')
best_customers_df = df_merged[df_merged['Segment'] == 'Best Customers']

plt.figure(figsize=(10, 6))
sns.histplot(best_customers_df['TransactionValue'], kde=True, color='green', bins=30)

plt.title('Transaction Value Distribution for "Best Customers"', fontsize=15, weight='bold')
plt.xlabel('Transaction Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```
